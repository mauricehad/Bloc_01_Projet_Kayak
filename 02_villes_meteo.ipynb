```python
# faire la même chose en boucle for pour toutes les destinations (ville_urls.csv)

villes_coordonnees = pd.read_csv("./csv_files/villes_coordonnees.csv")
huit_jours_toutes_villes = pd.DataFrame()

for _, row in villes_coordonnees.iterrows():
    lat = row['lat']
    lon = row['lon']
    id_destination = row['id_destination']
    destination = row['destination']
    
    url = "https://api.openweathermap.org/data/3.0/onecall?"
    params = {
        'lat': lat,
        'lon': lon,
        'exclude': "current,minutely,hourly,alerts",
        'units': "metric",
        'lang': "fr",
        'appid': api_openweather  # Assurez-vous que cette variable est définie avant
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        meteo_ville_unique = response.json()
        
        huit_jours_ville_unique = pd.DataFrame([
            {
                'date': datetime.datetime.fromtimestamp(day['dt']).strftime('%d-%m-%Y'),
                'id_destination': id_destination,
                'destination': destination,
                'description': day['weather'][0]['description'],
                'min_temp_C': day['temp']['min'],
                'max_temp_C': day['temp']['max'],
                'temp_ressentie_C': day['feels_like']['day'],
                'humidite_pourcent': day['humidity'],
                'vent_ms': day['wind_speed'],
                'indice_uv': day['uvi'],
                'pluie_probabilite': day['pop'] * 100,
                'pluie_mm': day.get('rain', 0),
                'nuageux_pourcent': day['clouds']
            }
            for day in meteo_ville_unique['daily']
        ])
        
        huit_jours_toutes_villes = pd.concat([huit_jours_toutes_villes, huit_jours_ville_unique], ignore_index=True)
        
    else:
        print(f"Erreur pour la destination {destination}, code HTTP : {response.status_code}")

huit_jours_toutes_villes
```

```python
# ajouter le scoring :

def calcule_score(row):
    # Température maximale
    if 18 <= row['max_temp_C'] <= 30:
        score_max_temp = 1
    elif row['max_temp_C'] < 18:
        score_max_temp = row['max_temp_C'] / 18
    else:
        score_max_temp = max(0, (40 - row['max_temp_C']) / 10)

    # Température minimale
    if 10 <= row['min_temp_C'] <= 18:
        score_min_temp = 1
    elif row['min_temp_C'] < 10:
        score_min_temp = row['min_temp_C'] / 10
    else:
        score_min_temp = max(0, (25 - row['min_temp_C']) / 7)

    # Température ressentie
    if 17 <= row['temp_ressentie_C'] <= 25:
        score_temp_ressentie = 1
    elif row['temp_ressentie_C'] < 17:
        score_temp_ressentie = max(0, (row['temp_ressentie_C'] - 5) / 12)
    else:
        score_temp_ressentie = max(0, (35 - row['temp_ressentie_C']) / 10)

    # Humidité
    if 40 <= row['humidite_pourcent'] <= 60:
        score_humidite = 1
    elif row['humidite_pourcent'] < 40:
        score_humidite = max(0, (row['humidite_pourcent'] - 20) / 20)
    else:
        score_humidite = max(0, (80 - row['humidite_pourcent']) / 20)

    # Vent
    if row['vent_ms'] <= 5:
        score_vent = 1
    elif 5 < row['vent_ms'] <= 8:
        score_vent = max(0, (8 - row['vent_ms']) / 3)
    else:
        score_vent = 0

    # Indice UV
    if 1 <= row['indice_uv'] <= 5:
        score_uv = 1
    elif row['indice_uv'] < 1:
        score_uv = row['indice_uv']
    else:
        score_uv = max(0, (10 - row['indice_uv']) / 5)

    # Probabilité de pluie
    if row['pluie_probabilite'] <= 20:
        score_prob_pluie = 1
    elif 20 < row['pluie_probabilite'] <= 80:
        score_prob_pluie = max(0, (80 - row['pluie_probabilite']) / 60)
    else:
        score_prob_pluie = 0

    # Quantité de pluie
    if row['pluie_mm'] <= 1:
        score_quant_pluie = 1
    elif 1 < row['pluie_mm'] <= 10:
        score_quant_pluie = max(0, (10 - row['pluie_mm']) / 9)
    else:
        score_quant_pluie = 0

    # Couverture nuageuse
    if 10 <= row['nuageux_pourcent'] <= 40:
        score_nuageux = 1
    elif row['nuageux_pourcent'] < 10:
        score_nuageux = row['nuageux_pourcent'] / 10
    elif 40 < row['nuageux_pourcent'] <= 80:
        score_nuageux = max(0, (80 - row['nuageux_pourcent']) / 40)
    else:
        score_nuageux = 0

    # Pondération des scores
    final_score = (
        (score_max_temp * 0.05) +
        (score_min_temp * 0.05) +
        (score_temp_ressentie * 0.20) +
        (score_humidite * 0.05) +
        (score_vent * 0.15) +
        (score_uv * 0.04) +
        (score_prob_pluie * 0.20) +
        (score_quant_pluie * 0.11) +
        (score_nuageux * 0.15)
    )

    return final_score

huit_jours_toutes_villes['scoring'] = huit_jours_toutes_villes.apply(calcule_score, axis=1)
huit_jours_toutes_villes
```

```python
# considérons 7 jours 

sept_jours_toutes_villes = huit_jours_toutes_villes.groupby('destination', group_keys=False).nth(slice(1,None)).reset_index(drop=True)
sept_jours_toutes_villes
```

```python
# exporter la meteo jour par jour de chaque ville en csv :

sept_jours_toutes_villes.to_csv('./csv_files/villes_meteo.csv', index=False)
```

```python
# Créer des colonnes pour la date d'arrivée et la date de départ
date_arrivee = sept_jours_toutes_villes["date"].min()
date_depart = sept_jours_toutes_villes["date"].max()

# Grouper par id_destination et calculer la moyenne pour toutes les colonnes numériques
villes_scoring_meteo = (
    sept_jours_toutes_villes.groupby(["id_destination", "destination"], as_index=False)
    .agg(
        scoring_sejour=("scoring", "mean"),
        min_temp_C_mean=("min_temp_C", "mean"),
        max_temp_C_mean=("max_temp_C", "mean"),
        temp_ressentie_C_mean=("temp_ressentie_C", "mean"),
        humidite_pourcent_mean=("humidite_pourcent", "mean"),
        vent_ms_mean=("vent_ms", "mean"),
        indice_uv_mean=("indice_uv", "mean"),
        pluie_probabilite_mean=("pluie_probabilite", "mean"),
        pluie_mm_mean=("pluie_mm", "mean"),
        nuageux_pourcent_mean=("nuageux_pourcent", "mean")
    )
)

# Ajouter les colonnes date_arrivee et date_depart
villes_scoring_meteo["date_arrivee"] = date_arrivee
villes_scoring_meteo["date_depart"] = date_depart

# Réorganiser les colonnes pour correspondre au format souhaité
villes_scoring_meteo = villes_scoring_meteo[
    [
        "date_arrivee", 
        "date_depart", 
        "id_destination", 
        "destination", 
        "scoring_sejour", 
        "min_temp_C_mean", 
        "max_temp_C_mean", 
        "temp_ressentie_C_mean", 
        "humidite_pourcent_mean", 
        "vent_ms_mean", 
        "indice_uv_mean", 
        "pluie_probabilite_mean", 
        "pluie_mm_mean", 
        "nuageux_pourcent_mean"
    ]
]

# Trier les destinations par scoring décroissant
villes_scoring_meteo = villes_scoring_meteo.sort_values(['scoring_sejour'], ascending=False)

# Afficher le résultat
villes_scoring_meteo
```

```python
# exporter le scoring du sejour par ville en .csv :

villes_scoring_meteo.to_csv('./csv_files/villes_scoring_meteo.csv', index=False)
```
